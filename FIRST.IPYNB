```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('matches.csv')
df1.shape
df1.head()
```

```python
df2 = pd.read_csv('deliveries.csv')
df2.shape
df2.head()
```

```python
total_score_df = df2.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
total_score_df.head()
```

```python
total_score_first = total_score_df[total_score_df['inning'] == 1]
total_score_first.head()
```

```python
df = df1.merge(total_score_first[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
df.head()
```

```python
df['team1'].unique()
```

```python
df['team1'] = df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
df['team1'].unique()
```

```python
df['team2'] = df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
df['team2'] = df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
df['team1'] = df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
df['team1'].unique()
```

```python
df['team1'] = df['team1'].str.replace('Royal Challengers Bengaluru','Royal Challengers Bengalore')
df['team2'] = df['team2'].str.replace('Royal Challengers Bengaluru','Royal Challengers Bengalore')
```

```python
df['team1'] = df['team1'].str.replace('Punjab Kings','Kings XI Punjab')
df['team2'] = df['team2'].str.replace('Punjab Kings','Kings XI Punjab')
```

```python
teams = ['Royal Challengers Bengalore','Kings XI Punjab','Delhi Capitals','Mumbai Indians','Kolkata Knight Riders', 'Rajasthan Royals',
       'Sunrisers Hyderabad', 'Chennai Super Kings','Gujarat Titans',
       'Royal Challengers Bengaluru']
```

```python
matches = df[df['team1'].isin(teams)]
matches = df[df['team2'].isin(teams)]
```

```python
match_df = matches[['match_id','city','winner','total_runs']]
```

```python
match_df.head()
```

```python
delivery_df = match_df.merge(df2,on='match_id')
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df.shape
```

```python
delivery_df.head()
```

```python
delivery_df['curr_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
delivery_df.head(10)
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x']-delivery_df['curr_score']+1; 
```

```python
delivery_df.head()
```

```python
delivery_df.head()

delivery_df['balls_left'] = 120-(delivery_df['over']*6 + delivery_df['ball'])
delivery_df.head()
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
delivery_df.head()
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_dismissed'].unique()
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
delivery_df['wickets_left'] = 11 - (delivery_df.groupby('match_id')['player_dismissed'].cumsum().values)
```

```python
delivery_df.head(10)
```

```python
#crr
delivery_df['crr'] = ((delivery_df['curr_score']*6) / (120-delivery_df['balls_left']))
delivery_df.head()
```

```python
delivery_df['rrr'] = ((delivery_df['runs_left'] *6)/ delivery_df['balls_left'])
delivery_df.head(15)
```

```python
delivery_df['city'].unique()
```

```python
def result(row):
   return 1 if row['batting_team'] == row['winner'] else 0


delivery_df['result'] = delivery_df.apply(result,axis = 1)
```

```python
delivery_df.columns
```

```python
final_df = delivery_df[['city','batting_team',
       'bowling_team', 'total_runs_x','runs_left', 'balls_left', 'wickets_left', 'crr', 'rrr',
       'result']]
```

```python
final_df = final_df.sample(final_df.shape[0])
final_df.head()
```

```python
final_df.isnull().sum()
final_df.dropna(inplace=True)
final_df['rrr'].describe()
final_df = final_df[final_df['balls_left'] != 0]
final_df['rrr'].describe()
```

MODEL TRAINING


```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1];

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)

X_train
```

```python
delivery_df[(delivery_df['batting_team'] == 'Kolkata Knight Riders') &( delivery_df['bowling_team'] == 'Royal Challengers Bangalore') ]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer(
    [
        ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
    ],
    remainder='passthrough'
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps = [
   ('step1',trf),
   ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
